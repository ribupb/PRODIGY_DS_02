# <h3>Distribution of the Target Variable</h3>
<p>Let's visualize the distribution of the target variable which indicates the presence of heart disease.</p>

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Target distribution
sns.countplot(x='target', data=heart_disease)
plt.title('Distribution of Heart Disease')
plt.xlabel('Presence of Heart Disease')
plt.ylabel('Count')
plt.show()
```

# <h3>Numerical Features</h3>
<p>We will visualize the distribution of key numerical features and their relationship with the target variable.</p>

```python
# Distribution of Age
sns.histplot(heart_disease['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Scatter plot of Age vs Maximum Heart Rate Achieved
sns.scatterplot(x='age', y='thalach', hue='target', data=heart_disease)
plt.title('Age vs. Maximum Heart Rate Achieved')
plt.show()
```

```python
# Box plot of Resting Blood Pressure by Heart Disease
sns.boxplot(x='target', y='trestbps', data=heart_disease)
plt.title('Resting Blood Pressure by Heart Disease')
plt.xlabel('Presence of Heart Disease')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

```python
# Box plot of Serum Cholesterol by Heart Disease
sns.boxplot(x='target', y='chol', data=heart_disease)
plt.title('Serum Cholesterol by Heart Disease')
plt.xlabel('Presence of Heart Disease')
plt.ylabel('Serum Cholesterol')
plt.show()
```

# <h3>Categorical Features</h3>
<p>Now let us examine the impact of some categorical features on the target variable.</p>

```python
# Heart Disease by Chest Pain Type
sns.countplot(x='cp', hue='target', data=heart_disease)
plt.title('Heart Disease by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python
# Heart Disease by Sex
sns.countplot(x='sex', hue='target', data=heart_disease)
plt.title('Heart Disease by Sex')
plt.xlabel('Sex (0 = female, 1 = male)')
plt.ylabel('Count')
plt.show()
```

```python
# Heart Disease by Fasting Blood Sugar
sns.countplot(x='fbs', hue='target', data=heart_disease)
plt.title('Heart Disease by Fasting Blood Sugar')
plt.xlabel('Fasting Blood Sugar > 120 mg/dl')
plt.ylabel('Count')
plt.show()
```

```python
# Heart Disease by Resting ECG
sns.countplot(x='restecg', hue='target', data=heart_disease)
plt.title('Heart Disease by Resting ECG')
plt.xlabel('Resting ECG')
plt.ylabel('Count')
plt.show()
```

```python
# Heart Disease by Exercise-Induced Angina
sns.countplot(x='exang', hue='target', data=heart_disease)
plt.title('Heart Disease by Exercise-Induced Angina')
plt.xlabel('Exercise-Induced Angina')
plt.ylabel('Count')
plt.show()
```

```python
# Heart Disease by Slope of Peak Exercise ST Segment
sns.countplot(x='slope', hue='target', data=heart_disease)
plt.title('Heart Disease by Slope of Peak Exercise ST Segment')
plt.xlabel('Slope of Peak Exercise ST Segment')
plt.ylabel('Count')
plt.show()
```

```python
# Heart Disease by Number of Major Vessels
sns.countplot(x='ca', hue='target', data=heart_disease)
plt.title('Heart Disease by Number of Major Vessels')
plt.xlabel('Number of Major Vessels (0-3) colored by fluoroscopy')
plt.ylabel('Count')
plt.show()
```

```python
# Heart Disease by Thalassemia
sns.countplot(x='thal', hue='target', data=heart_disease)
plt.title('Heart Disease by Thalassemia')
plt.xlabel('Thalassemia (3 = normal, 6 = fixed defect, 7 = reversible defect)')
plt.ylabel('Count')
plt.show()
```

# <h3>Correlation Analysis</h3>
<p>Let's look at the correlation between numerical features and the target variable.</p>

```python
# Correlation matrix
corr_matrix = heart_disease.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

# Strongest correlations with Target
corr_target = corr_matrix['target'].sort_values(ascending=False)
print(corr_target.head(10))
print(corr_target.tail(10))
```
